```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'EmployeeID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eva', 'Frank', 'Grace', 'Henry'],
    'Department': ['HR', 'Engineering', 'Marketing', 'Finance', 'HR', 'Engineering', 'Marketing', 'Finance'],
    'DateJoined': ['2020-01-15', '2019-06-01', '2021-03-10', '2018-11-20', '2022-02-05', '2020-08-17', '2019-09-10', '2021-12-01'],
    'DateLeft': ['2023-02-28', None, '2022-12-15', None, None, '2023-03-01', None, None],
    'Salary': [60000, 80000, 55000, 90000, 62000, 85000, 58000, 92000]
}
df = pd.DataFrame(data)
df['DateJoined'] = pd.to_datetime(df['DateJoined'])
df['DateLeft'] = pd.to_datetime(df['DateLeft'])

df
```

```python
df['Status']=df['DateLeft'].apply(lambda x: 'Former' if pd.notnull(x) else 'Current')
df[['EmployeeID', 'Name', 'Department', 'Status']]
```

```python
status_count=df['Status'].value_counts()
print(status_count)
```

```python
status_count.plot(kind='bar',color=['green','red'])
plt.title('Employee Status Count')
plt.xlabel('Status')
plt.ylabel('Number of Employees')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
dept_counts=df['Department'].value_counts()
print(dept_counts)
```

```python
dept_counts.plot(kind='bar',color='blue')
plt.title('Employees Per department')
plt.xlabel('Department')
plt.ylabel('Number of Emplooyees')
plt.grid(True)
plt.show()
```

```python
dept_status_counts=df.groupby(['Department','Status']).size().unstack()
dept_status_counts.plot(kind='bar',stacked=True,colormap='Paired')
plt.title('Current VS Former Employees by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.grid(True)
plt.show()
```

```python
avg_salary_by_dept=df.groupby('Department')['Salary'].mean().sort_values(ascending=False)
print(avg_salary_by_dept)
avg_salary_by_dept.plot(kind='bar',color='green')
plt.title('Average Salary By Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
df['Tenure']=(df ['DateLeft'].fillna(pd.Timestamp('today')) - df['DateJoined']).dt.days // 365
df[['EmployeeID','Name','DateLeft','DateJoined','Tenure']]
print(df)
avg_tenure_by_department=df.groupby('Department')['Tenure'].mean().sort_values(ascending=False)
print(avg_tenure_by_department)
avg_tenure_by_department.plot(kind='bar',color='brown')
plt.title('Average Employee Year By Department')
plt.xlabel('Department')
plt.ylabel('Average Tenure (Years)')
plt.grid(True)
plt.show()
```
